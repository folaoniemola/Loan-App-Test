```python
## to split our data into training and test data
from sklearn.model_selection import train_test_split

## to check the accuracy of the model
from sklearn.metrics import accuracy_score 
```

```python
## renaming the loan status to targert column

df_missing.rename(columns = {'Loan_Status': 'Target'}, inplace = True)
```

```python
df_missing
```

```python
df_missing['Property_Area'].value_counts()
```

```python
#df_missing['gender'] = df_missing['Gender'].replace({'Female':0, 'Male':1}).astype(int)
```

```python
pd.set_option('future.no_silent_downcasting', True)
df_missing['Gender'] = df_missing['Gender'].replace({'Female':0, 'Male':1}).astype(int)
df_missing['Married'] = df_missing['Married'].replace({'No':0, 'Yes':1}).astype(int)
df_missing['Education'] = df_missing['Education'].replace({'Not Graduate':0, 'Graduate':1}).astype(int)
df_missing['Self_Employed'] = df_missing['Self_Employed'].replace({'No':0, 'Yes':1}).astype(int)
```

```python
df_missing
```

```python
df_missing['Gender'].value_counts()
```

```python
## One Hot Encoding

df_dummies = pd.get_dummies(df_missing, columns = ['Property_Area'])
df_dummies = df_dummies.drop(columns = ['Property_Area_Semiurban'])## Dropping one of the columns
```

```python
df_dummies
```

```python
pd.set_option('future.no_silent_downcasting', True)
df_dummies['Property_Area_Rural'] = df_dummies['Property_Area_Rural'].replace({'False':0, 'True':1}).astype(int)
df_dummies['Property_Area_Urban'] = df_dummies['Property_Area_Urban'].replace({'False':0, 'True':1}).astype(int)
```

```python
df_dummies
```

```python
df_dummies['Target'] = df_dummies['Target'].replace({'N':0, 'Y':1}).astype(int)
```

```python
df_dummies.head()
```

```python
## Saving cleaned data to python desktop

## df_dummies.to_csv('cleaned_data.csv', index = False)
```

```python
## replacing 3+ in the dependents column with 4
df_dummies.replace('3+', 4, inplace = True)
```

```python
## Separating the feature from the target

x = df_dummies.drop(columns = 'Target', axis = 1)
y = df_dummies['Target']
```

```python
## importing library
from sklearn.preprocessing import StandardScaler
```

```python
## Data Standardization; its the process of converting our values to a common range for our model to be able to predict
scaler = StandardScaler()
```

```python
scaler
```

```python
## fit and transform is performed only on the training data to avoid data leakage
```

```python
x_train
```

```python
x_test_scaled = scaler.transform(x_test)
```

```python
x_test_scaled
```

```python
## for classification model, we can use support vector machine, logistics, Random Forest, XGBoost
```

```python
# Classification Algorithm
#from sklearn import svm
from sklearn import svm
```

```python
# svm kernel 
classifier = svm.SVC(kernel = 'linear')
```

## Model Evaluation

```python
## predicting the training data accuracy of the model
x_train_prediction = classifier.predict(x_train_scaled)
training_data_accuracy = accuracy_score(x_train_prediction, y_train)
```

```python
## printing the score

print('Training Data Accuracy Score:', training_data_accuracy)
```
